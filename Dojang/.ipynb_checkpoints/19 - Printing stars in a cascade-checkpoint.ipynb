1. 중첩 루프 사용하기 (Using nested loops)

이걸 range()로 표현하면?     
0부터 4까지니 range(5)가 된다.

바깥쪽 루프 (세로) 우선 카운팅하면 range(5),
안쪽 루프 (가로)를 그 다음 카운팅해도 range(5)

```python
for i in range(5):                         # 5번 반복. 바깥쪽 루프는 세로 방향
    for j in range(5):                     # 5번 반복. 안쪽 루프는 가로 방향
        print('j:', j, sep='', end=' ')    # j값 출력. end에 ' '를 지정하여 줄바꿈 대신 한 칸 띄움
    print('i:', i, '\\n', sep='')          # i값 출력, 개행 문자 모양도 출력
                                           # 가로 방향으로 숫자를 모두 출력한 뒤 다음 줄로 넘어감
                                           # (print는 기본적으로 출력 후 다음 줄로 넘어감)
```

*복습* | 기억해야 할 것:
1. range는 0부터 카운트하기 때문에 (0,1,2,3,4)일때 총 range(5)가 된다.
2. sep=''는 요소 사이마다 무엇이 들어갈지 정하는거고 end=''는 print()안에 들어가는 합체 요소 끝에 무엇이 들어갈지 정하는 것이다.
3. '\n'을 문자열로 표현하기 위해서 \ reverse slash를 앞에 놓아준다. 즉 \n을 표현하려면 '\\n' 이렇게 써주면 된다.
4. end=''를 쓰면 줄바꾸기를 한줄로 바꾸니까, 바깥쪽 루프 (세로 방향)을 나타낼때 쓰지 않는다.

```python
# 따라쓰기 연습
for i in range(5):                        # repeat 5 times vertically
    for j in range(5):                    # repeat 5 times horizontally
        print('j:', j, sep='',end=' ')    # print 'j:'j horizontally
    print('i:', i, '\\n', sep='')         # print 'i:'i'\n' vertically
```

```python
for i in range(5):
    for j in range(5):
        print('*', end='')
    print()                   # 세로 가로 같은 문자열이기에 print()로 세로 줄 바꿔주기만 적용한다.                    
```

```python
for i in range(5):
    for j in range(5):
        print('k','y','u','r','i', sep='', end=' ')
    print()
    
```

```python
for i in range(6):
    for j in range(3):
        print('👽', end=' ')
    print('🚀', end=' ')
```

```python
for i in range(6):                   # 6줄의 로켓과 로켓줄 당 세개의 외계인
    for j in range(3):
        print('👽', end=' ')         # 6 x 3의 네모
    print('🚀')
```

2. 계단식으로 별 출력하기 (printing stars in cascade)

```python
for i in range(5):
    for j in range(5):
        #여기서 우린 if 조건문을 써야된다. 왜냐하면 j가 0일때부터 j가 0,1,2,3,4 일때까지 i의 범위만큼 출력되기 때문이다.
        if j <= i:
        #j가 i보다 작거나 같을때. i가 0일때 j가 0이면 작거나 같다 - 출력. i가 1일때 j가 (0,1)이면 작거나 같다. 출력. i가 4일때까지 반복
            print('*', end='')
    print()
```

```python
for i in range(5):
    for j in range(8):
        if i >= j:
            print('🐍', end='')
    print()
```

when i is 0,   
j is 0,   
when i is 1,   
j is 0, j is 1   
when i is 2,   
j is 0, j is 1, j is 2   

and so on   

```python
for i in range(5):
    for j in range(5):
        if i == j:
            print('🍇', end='')
        else:
            print('  ', end='')
    print()
```

**기억하기**       
우선 바깥쪽 루프가 세로 방향, 안쪽 루프가 가로 방향을 처리한다는 점만 기억하면 된다!

---

```python
for i in range(5):
    for j in range(5):
               
#       i(0)
#       j(0,1,2,3,4)
        
#       i(0,1)
#       j(0,1,2,3) 공백 하나
        
#       i(0,1,2)
#       j(0,1,2) 공백 둘
        
#       i(0,1,2,3)
#       j(0,1) 공백 셋
    
#       i(0,1,2,3,4)
#       j(0) 공백 넷

        if i < j:
            print('*', end='')
        else:
            print(' ', end='')
    print('*')
```

표준 입력으로 삼각형의 높이가 입력됩니다.    
입력된 높이만큼 산 모양으로 별을 출력하는 프로그램을 만드세요(input에서 안내 문자열은 출력하지 않아야 합니다).    
이때 출력 결과는 예제와 정확히 일치해야 합니다. 모양이 같더라도 공백이나 빈 줄이 더 들어가면 틀린 것으로 처리됩니다.

```python
# 우선 높이는 세로이기 때문에 i이다. i의 범위 숫자가 입력됨으로 범위를 변수 x로 지정해주자. 

x = int(input())

for i in range(x):
#             0   1  2  3   4
#    i(0), j(공백,공백,2,공백,공백)      i가 j랑 같고, 더 클때
#             0         4
#    i(1), j(공백,1,2,3,공백)          i가 j보다 크다 
#
#    i(2), j(0,1,2,3,4)
    
    for j in reversed(range((x*2)-1)):

        if i <= j :
            print('', end='')
        else:
            print('*', end='')
        if j == 2:
            print('*', end='')
        else:
            print('', end='')
        if j == (i+-2):
            print('*', end='')
            
    print()
```
