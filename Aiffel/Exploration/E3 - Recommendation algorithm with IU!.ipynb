## Preprocessing to utilise model (Practice)
---
**indexing** to manage data at ease

pandas.DataFrame.unique() is a very useful method which collects unique data in specific column.    
especially during the process of **indexing**

```python
# code for finding native users & artists
user_unique = data['user_id'].unique()
artist_unique = data['artist'].unique()

# indexing users & artists
# idx is short for index
user_to_idx = {v:k for k,v in enumerate(user_unique)}
artist_to_idx = {v:k for k,v in enumerate(artist_unique)}

# this is called dictionary, which comprises of keys and values
```

```python
# checking if indexing is done properly
print(user_to_idx['zimin']) # 358868th user
print(artist_to_idx['jason mraz']) #3746th artist
```

```python
temp_user_data = data ['user_id'].map(user_to_idx.get).dropna()
if len(temp_user_data) == len(data):
    print('user_id column has been indexed succesfully 😋')
    data['uder_id'] = temp_user_data
else:
    print('user_id column has failed at indexing 😞')
    
temp_artist_data = data['artist'].map(artist_to_idx.get).dropna()
if len(temp_artist_data) == len(data):
    print('user_id column has been indexed succesfully 😋')
else:
    print('user_id column has failed at indexing 😞')

# don't forget data to show the table!    
data 
```
