## 5-7. 미분과 머신러닝 (2) 파이썬 미분 구현



위 식을 보고 어떤 함수 f를 x=var에서 미분한 미분계수 $$f'(var)를 구하는 코드를 작성해 봅시다.

```python
def derivative(f, var):
    h = 1e-5
    der = (f(var + h) - f(var-h)) /( 2*h)
    return der
```

```python
def func(x):
    return x**2

derfunc = derivative(func, 3)

derfunc
```
