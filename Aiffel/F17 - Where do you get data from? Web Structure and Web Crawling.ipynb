# 17-4. Internet and Protocol (3) Socket communication

---

> In computer networking, an Internet socket, or network socket, means an **endpoint**.

> Among the transport areas, the protocol that implements **the peer-to-peer connection** between sockets to sockets is <mark>TCP</mark>.

> These sockets are numbered. This is called the **port number**. 

As we'll cover in more detail in the next steps, it is customarily promised to connect to port 80 of the web server when a web browser connects to a web server. The port that provides the HTTP protocol service is set to 80. In addition, there are conventionally promised port numbers such as 21 for FTP, 22 for SSH, and 23 for telnet.



```python
import socket
# open one socket
with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
    s.connect(('www.google.com', 80))      # connect to the host called google.com, the web host's number 80 port
    cmd='Get http://www.google.com HTTP/1.0\r\n\r\n'.encode()
    s.send(cmd)                            # once connected, send order of HTTP protocol in GET xxxxx format
    data=s.recv(1024)                      # And receive an HTTP response of 1024 bytes.
# let us check the respone
print('Received', data)
```

Let's learn about web before going into studying API.

The concept of API was introduced by Tim Berners-Lee. He suggested:    

> "When exchanging information on the Internet, let's connect information with hypertext!" 

Therefore, the document, protocol and identification rules of the web are as follows.

- Expression format(document): HTML(Hyper Text Markup Language)
- Sending method(protocol): HTTP (Hyper Text Transfer Protocol)
- Identifier: URI (Uniform Resource Identifier)

### Web behavior and HTTP 
---
The web uses HTTP as a **communication protocol**, and HTTP follows a server/client model that makes a request and responds.

### HTTP message
---
HTTP message comprises of **request/status line, header and body.**
The request line containing the request content includes the **request method and target address.**

#### Request Method (요청 메소드)

- **GET**: read, information request
- **HEAD**: only head info request
- **POST**: create and used to create the info, or transform the info
- **PUT**: update, edit information
- **DELETE**: delete, delete information
- **OPTIONS**: confirm usable methods

### Response Code (응답 코드)

* 1xx: Conditional Response   
* 2xx: Success    
* 3xx: Redirection Complete
* 4xx: Request Error(Something wrong with the Client!)    
* 5xx: Server Error(Problem with the server!)    

#### Response Code that are used often

- 200: the server completes client request successfuly & returns if page request is successfully complete in web pages
- 404: unable to find. This is a response code when inexisting URL is requested & cannot connect to the requested page

### how to use urllib

```python
# 1. webpage download

import urllib
def download(url):
    return urllib.request.urlopen(url)

# use the urlopen() function to convert the object of "HTTP Response" from the requested server.
# the problem with this function is that when you are trying to download the webpage, there's an error.

from urllib.error import URLError, HTTPError, ContentTooShortError

def download(url):
    try:
        html = urllib.request.urlopen(url)
    except (URLError, HTTPError, ContentTooShortError) as e:
        print('Download error', e.reason)
        html = None
    return html

download('https://www.google.com')
```

With urlopen() function, you can only open url in http protocol. For details, refer to the urllib.request specification. Checks the occurrence of an exception and adds an exception handling code.

https://docs.python.org/3/library/urllib.request.html#module-urllib.request

```python
def download(url):
    try:
        html = urllib.request.urlopen(url).read()
    except (URLError, HTTPError, ContentTooShortError) as e:
        print('Download Error', e.reason)
        html = None
    return html
```
