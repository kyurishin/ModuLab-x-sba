# 19-2. Implementation and Sample Space & Events
---
> Experiments are the process of trial and observation.
> A random experiment: an experiment in which the results are <mark>unknown</mark> in some experiments that **can be repeated several times under the same conditions**, but <mark>all possible results</mark> can be known in advance.

1. If you can throw a dice under the same condition, like with the same strength of wind, the angle and speed in which the dice is thrown at; It is referred as an experiment that "can be repeated several times under the same conditions."
2. If it's uncertain which number will come out when the dice is thrown, it means "an experiment with unknown results"
3. If you can predict which number will come out, it means "a forseeable experiment". Therefore you can predict which ever possible numbers that can come out. {1, 2, 3, 4, 5, 6}

```python
import matplotlib.pyplot as plt

from matplotlib_venn import venn2, venn2_circles, venn3, venn_circles

print('1=1')
```

```python
# 교집합, intersection

fig = plt.figure()
fig.set_size_inches(10,8)

Num_Set_A = 100            # A 집합의 원소 개수입니다. 마음대로 바꾸실 수 있어요.
Num_Set_B = 70             # B 집합의 원소 개수입니다. 마음대로 바꾸실 수 있어요.
Num_Set_A_and_B = 20        # A와 B의 교집합의 원소 개수입니다. 마음대로 바꾸실 수 있어요.

v = venn2(subsets={'10': Num_Set_A, '01': Num_Set_B, '11': Num_Set_A_and_B}, set_labels = ('A', 'B')) 
c = venn2_circles(subsets=(Num_Set_A, Num_Set_B, Num_Set_A_and_B), linestyle='solid') 
v.get_patch_by_id('10').set_alpha(0.8) 

v.get_patch_by_id('10').set_color('w') 
v.get_patch_by_id('01').set_color('w') 
v.get_patch_by_id('11').set_color('#000034') 

v.get_label_by_id('10').set_text(' ') 
v.get_label_by_id('01').set_text(' ') 
v.get_label_by_id('11').set_text(' ') 

v.get_label_by_id('A').set_text('Set A') 
v.get_label_by_id('B').set_text('Set B') 
plt.gca().set_facecolor("white")
plt.gca().set_axis_on()
```

```python
# 합집합, union

fig = plt.figure()
fig.set_size_inches(10,8)

Num_Set_A = 13            # A 집합의 원소 개수입니다. 마음대로 바꾸실 수 있어요.
Num_Set_B = 8             # B 집합의 원소 개수입니다. 마음대로 바꾸실 수 있어요.
Num_Set_A_and_B = 5       # A와 B의 교집합의 원소 개수입니다. 마음대로 바꾸실 수 있어요.

v = venn2(subsets={'10': Num_Set_A, '01': Num_Set_B, '11': Num_Set_A_and_B}, set_labels = ('A', 'B')) 
c = venn2_circles(subsets=(Num_Set_A, Num_Set_B, Num_Set_A_and_B), linestyle='solid') 
v.get_patch_by_id('10').set_alpha(0.8) 

v.get_patch_by_id('10').set_color('lightyellow') 
v.get_patch_by_id('01').set_color('aquamarine') 
v.get_patch_by_id('11').set_color('darkblue') 

v.get_label_by_id('10').set_text(' ') 
v.get_label_by_id('01').set_text(' ') 
v.get_label_by_id('11').set_text(' ') 

v.get_label_by_id('A').set_text('Set A') 
v.get_label_by_id('B').set_text('Set B') 
plt.gca().set_facecolor("white")
plt.gca().set_axis_on()
```

```python
# 서로소 집합, disjoint sets

fig = plt.figure()
fig.set_size_inches(10,8)

Num_Set_A = 13             # A 집합의 원소 개수입니다. 마음대로 바꾸실 수 있어요.
Num_Set_B = 8               # B 집합의 원소 개수입니다. 마음대로 바꾸실 수 있어요.
Num_Set_A_and_B = 5    # A와 B의 교집합의 원소 개수입니다. 마음대로 바꾸실 수 있어요.

v = venn2(subsets={'10': Num_Set_A, '01': Num_Set_B}, set_labels = ('Set A', 'Set B'))
c = venn2_circles(subsets=(Num_Set_A, Num_Set_B, 0), linestyle='solid')

v.get_patch_by_id('10').set_alpha(0.5)
v.get_patch_by_id('10').set_color('r')
v.get_patch_by_id('01').set_alpha(0.5)
v.get_patch_by_id('01').set_color('gray')
v.get_label_by_id('10').set_text('')
v.get_label_by_id('01').set_text('')
v.get_label_by_id('A').set_text('A')
v.get_label_by_id('B').set_text('B')

plt.gca().set_axis_on()
```
