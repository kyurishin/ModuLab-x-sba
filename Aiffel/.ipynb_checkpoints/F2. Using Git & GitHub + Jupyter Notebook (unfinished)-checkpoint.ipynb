# 2-1. Learning Obj. and Contents

---

## Learning Premises<br> 
<font size="2">
    
- Never used collaborating tools for programming before.    
    
- Don't know what Git and GitHub are.    
    
- Never used GitHub to manage source codes' versions.     
    
- Familiar with using pip but never installed and ran Jupyter Notebook before.     
    
- Never studied / used the Markdown Syntax.         
</font>   


---

## Learning Objectives<br>
<font size="2">
    
- Able to tell the difference between Git and GitHub.     
    
- Understands and can utilise Git's commands: add, commit, push, pull.    
    
- Able to manage source codes' versions using GitHub.       
    
- Can code/write in Markdown using Jupyter Notebook.    
</font>     

---

## Contents<br>
<font size="2">    
    
1. The first Survival Skill for Developers              
    - Learn about the necessary tools for collaboration.
    - Understanding GitHub and its purpose.     
    
    
2. Can I record all the footsteps of my code?
    - Knowing the differences between Git and GitHub
    - Start installing them on Command Line Interface (CLI)
    
    
3. Plant your first lawn on GitHub, it's not difficult!
    - To use GitHub for the first time, create a local repository and a remote repository.
    - Try using commands such as: add, commit, push, and pull through hands-on practice, and know what they do.   


4. Mastering Jupyter Notebook, the MS word in data analytics
    - Observe what Jupyter Notebook is, and its use.
    - Learn the manual of Jupyter Notebook.   


5. Start and end of developer documentation! Let's learn Markdown
    - Learn various Markdown Syntax & Grammar
    - Memorise with hands-on practice!       
</font>
